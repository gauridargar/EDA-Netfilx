```python
tag = "Stand-Up Comedy" #standup shows on Netflix
df["relevant"] = df['listed_in'].fillna("").apply(lambda x : 1 if tag.lower() in x.lower() else 0)
comedy = df[df["relevant"] == 1]
comedy[comedy["country"] == "United States"][["title", "country","release_year"]].head(10)
```

```python
tag = "Kids' TV"
df["relevant"] = df['listed_in'].fillna("").apply(lambda x : 1 if tag.lower() in x.lower() else 0)
com = df[df["relevant"] == 1]
com[com["country"] == "United States"][["title", "country","release_year"]].head(10)
```

```python
plt.figure(figsize=(15,7))
df[df['type']=="Movie"]["release_year"].value_counts()[:15].plot(kind="bar", color ="Green")
plt.title("Freq of movies released in different years on Netflix")
plt.show()
```

```python
plt.figure(figsize=(15,7))
df[df["type"] == "TV Show"]["release_year"].value_counts()[:15].plot(kind="bar", color="Orange")
plt.title("Freq of TV Shows released in different years on Netflix")
plt.show()
```

```python
plt.figure(figsize=(15,7))
df[df["type"]=="Movie"]["listed_in"].value_counts()[:15].plot(kind="bar", color="Red")
plt.title("Top 15 Genre for Moives on Netflix")
plt.show()
```

```python
plt.figure(figsize=(15,7))
df[df["type"]=="TV Show"]["listed_in"].value_counts()[:15].plot(kind="bar", color="Blue")
plt.title("Top 15 Genre for of TV Shows on Netflix")
plt.show()
```

```python
plt.figure(figsize=(15,7))
df.country.value_counts().head().sort_values().plot(kind="bar",color="Teal")
plt.title("Top Five Countries with the most Content on Netflix",fontsize=(18))
plt.xlabel("Country",fontsize=(15))
plt.ylabel("Amount of TV & Film Contents available on Netflix")
plt.show()
```

```python
plt.figure(figsize=(15,7))
df.listed_in.value_counts()[:10].plot(kind="bar",color=['#9FE2BF', '#DFFF00', '#FFBF00', '#FF7F50','#CCCCFF' ])
plt.title("Top Gener for Movies and TV Shows on Netflix")
plt.show()
```
